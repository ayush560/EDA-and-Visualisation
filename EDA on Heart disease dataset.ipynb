```python
num_cols = ['Age', 'RestingBloodPressure', 'Cholestrol', 'MaxHeartRateAchivied', 'Oldpeak']
categ_cols = ['Gender', 'Chestpain', 'FastingBloodSugar', 'RestingECG', 'ExerciseIndusedAngina', 'Slope', 'MajorVessels', 'Thalassemia']
```

```python
# let's check the distribution of each numerical attribute
for cols in num_cols:
    plt.figure()
    sns.distplot(data[cols].dropna())
    plt.xlabel(cols)
    plt.show()
```

### here we can see that all attributes are closer to normal distribution except Oldpeak

```python
# Now let's check whether distribution of numerical attributes plays any significant role in predicting heart disease or not
for cols in num_cols:
    plt.figure()
    sns.barplot(x=data['Target'],y=data[cols])
    plt.show()
```

## Takeaway Points:
### 1. Interestingly people having high age are less likely to suffer from heart disease, but the records measured here are random and vey few in numbers so we can't make any strong conclusion from it
### 2. People having more maximum heart rate are more likely to be affected from heart disease
### 3. People having high value of oldpeak are less likely o be affected from heart disease

```python
# Now, let's check on whether chestpain is playing an role in having heart disease or not
pd.crosstab(data['ChestPain'], data['Target']).plot(kind='bar')
plt.show()
```

```python
# do a gender comparison 
ax = sns.countplot(hue='ChestPain',x='Gender',data=data,palette='husl')

plt.title("Chest Pain Type Vs Gender")    
plt.ylabel("")
plt.yticks([])
plt.xlabel("")
for p in ax.patches:
    ax.annotate(p.get_height(),(p.get_x()+0.05, p.get_height()+1))
ax.set_xticklabels(['Female','Male'])
print(ax.patches)
plt.show()
```

### Here, it can be clearly inferred that :
#### 1. People having chest pain are more likely to be affected by heart diseases
#### 2. About 50% of total males suffer from Type0 Chest pain
